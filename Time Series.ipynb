```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('StockX-Data-Contest-2019-3.csv')
```

```python
sp = [float(x[1:].replace(',', '')) for x in list(df['Sale Price'])]
df['Sale Price'] = sp
rp = [float(x[1:].replace(',', '')) for x in list(df['Retail Price'])]
df['Retail Price'] = rp
```

```python
yzy = df[df['Brand'] == ' Yeezy']
```

```python
yzy_b = yzy[yzy['Sneaker Name'] == 'Adidas-Yeezy-Boost-350-V2-Beluga-2pt0']
yzy_b.drop_duplicates(subset=['Order Date'], ignore_index=True, inplace=True)
```

```python
yzy_b['Order Date'] = pd.to_datetime(yzy_b['Order Date'])
```

```python
yzy_b
```

```python
yzy_b = yzy_b.set_index('Order Date')
```

```python
yzy_b['Year'] = yzy_b.index.year
yzy_b['Month'] = yzy_b.index.month
yzy_b['Weekday Name'] = yzy_b.index.day_name()
```

```python
yzy_b
```

```python
sns.set(rc={'figure.figsize':(5, 10)})
plt.rcParams['figure.figsize'] = (8,5)
plt.rcParams['figure.dpi'] = 150
```

```python
yzy_b['Sale Price'].plot(linewidth=0.5)
```

```python
ax = yzy_b.loc['2018', 'Sale Price'].plot()
ax.set_ylabel('Sales Price')
```

```python
ax = yzy_b.loc['2018-07', 'Sale Price'].plot(marker='o', linestyle='-')
ax.set_ylabel('Sales Price')
```

```python
yzy_b['Return'] = yzy_b['Sale Price'] - yzy_b['Retail Price']
```

```python
fig, axes = plt.subplots(2, 1, figsize=(8, 7), sharex=True)

for name, ax in zip(['Sale Price', 'Return'], axes):
    sns.boxplot(data=yzy_b, x='Month', y=name, ax=ax)
    ax.set_ylabel('$')
    ax.set_title(name)
    if ax != axes[-1]:
        ax.set_xlabel('')
```

```python
sns.boxplot(data=yzy_b, x='Weekday Name', y='Sale Price')
```

```python
# Resampling by weekly mean

yzy_b_sp_w_m = yzy_b['Sale Price'].resample('W').mean()
yzy_b_sp_w_m
```
