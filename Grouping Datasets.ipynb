```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('StockX-Data-Contest-2019-3.csv')
```

```python
sp = [float(x[1:].replace(',', '')) for x in list(df['Sale Price'])]
df['Sale Price'] = sp
rp = [float(x[1:].replace(',', '')) for x in list(df['Retail Price'])]
df['Retail Price'] = rp
```

# Data Aggregation
Aggregation -> apply any mathematical operation on a dataset

Sum, min, max...

```python
infos = df[['Sale Price', 'Retail Price', 'Shoe Size']]

infos.agg('mean', axis='rows')
```

```python
infos.agg(['sum', 'min', 'max'])
```

```python
infos.agg({'Sale Price': ['sum', 'min', 'max'],
          'Retail Price': ['sum', 'min', 'max'],
          'Shoe Size': ['min', 'max']})
```

# Aggregation

```python
aggreg = {'Sale Price': 'mean', 'Shoe Size': [max,min]}
df.groupby(['Brand', 'Sneaker Name']).agg(aggreg)
```

```python
df.groupby(['Brand', 'Sneaker Name']).agg(max_price=('Sale Price', max), min_price=('Sale Price', min), avg_price=('Sale Price', 'mean'))
```

```python
df.groupby(['Brand', 'Sneaker Name'])['Sale Price'].transform(lambda x: x*5.34) #converting sale prices to BRL
```

```python
df['BRL_SP'] = df.groupby(['Brand', 'Sneaker Name'])['Sale Price'].transform(lambda x: x*5.34) 
```

```python
df[['Sale Price', 'BRL_SP']]
```

# Pivot tables

```python
table = pd.pivot_table(df, index=['Sneaker Name'])
```

```python
table
```

```python
table = pd.pivot_table(df, index=['Brand', 'Sneaker Name'])
```

```python
table
```

```python
pivot_df = df.filter(['Brand', 'Sneaker Name', 'Sale Price', 'Shoe Size'])
```

```python
table = pd.pivot_table(pivot_df, values='Sale Price', index=['Brand', 'Sneaker Name'],
                      columns=['Shoe Size'], aggfunc=np.mean, fill_value=0)
```

```python
table
```

```python
table[table.index.get_level_values('Brand') == ' Yeezy']
```

# Cross-tabulation

```python
pd.crosstab(df['Brand'], df['Shoe Size'], margins = True, margins_name='Total Shoes')
```
