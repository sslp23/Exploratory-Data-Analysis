```python
df = pd.read_csv('StockX-Data-Contest-2019-3.csv')
```

# Binning and Discretization
convert continuous variables in discrete intervals

```python
yzy_ = yzy_p_nd.copy()
yzy_['Profit'] = yzy_['Sale Price'] - yzy_['Retail Price']
```

```python
yzy_pos = yzy_[yzy_['Profit']>=0]
```

```python
bins = list(range(0, 2400, 200))
cats = pd.cut(yzy_pos['Profit'], bins)
```

```python
yzy_pos['Profit_Cat'] = cats.values
```

```python
# ( -> this side is open (extremes not included)
# [ -> this side is closed (inclusive)
yzy_pos
```

```python
## adding bins_labels
yzy_posl = yzy_[yzy_['Profit']>=0]
bins = list(range(0, 2800, 400))
labels = ('Average Profit', 'Good Profit', 'High Profit', 'Very High Profit', 'Amazing Profit',  'Insane Profit')
cats = pd.cut(yzy_pos['Profit'], bins, labels=labels)
```

```python
yzy_posl['Profit_Cat'] = cats.values
```
